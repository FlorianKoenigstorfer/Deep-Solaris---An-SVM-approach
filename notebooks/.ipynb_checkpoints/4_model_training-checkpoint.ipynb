## Introduction

This notebook implements a **HOG + SVM** binary classification pipeline to detect whether a satellite image tile contains a solar panel installation.

### Three sub-questions investigated

1) Which SVM **kernel function** is optimal?
2) Which HOG **cell/block size** combination is optimal?
3) Which number of HOG **orientation bins** is optimal?

### Methodology

All cross-validation uses a `Pipeline(StandardScaler, SVC)` so the scaler is fitted **only on the training folds** of each split — preventing data leakage from the validation fold into feature normalisation. Five-fold stratified CV is used throughout.

## Setup & Imports

```python
%matplotlib inline
import os
import sys
import warnings

from sklearn.exceptions import ConvergenceWarning

warnings.filterwarnings("ignore", category=ConvergenceWarning)
warnings.filterwarnings("ignore", category=FutureWarning)
warnings.filterwarnings("ignore", category=UserWarning, module="pandas")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display

sys.path.insert(0, "../src")
from model import (
    train_svm,
    cross_validate_svm,
    evaluate_on_holdout,
    plot_roc_curve,
    plot_precision_recall_curve,
    plot_confusion_matrix,
    plot_learning_curve,
    run_kernel_grid_search,
    run_hog_ablation,
)
from feature_extraction import load_feature_matrix, make_ablation_npz_name

FEATURES_DIR        = "../data"
FEATURES_TRAIN_PATH = "../data/features_train.npz"
FEATURES_TEST_PATH  = "../data/features_test.npz"
CV_FOLDS            = 5
RANDOM_STATE        = 42
BLOCK_NORM          = "L1"

if not os.path.exists(FEATURES_TRAIN_PATH):
    raise FileNotFoundError(
        f"Training features not found at '{FEATURES_TRAIN_PATH}'. "
        "Please run feature_extraction.ipynb first."
    )
if not os.path.exists(FEATURES_TEST_PATH):
    raise FileNotFoundError(
        f"Test features not found at '{FEATURES_TEST_PATH}'. "
        "Please run feature_extraction.ipynb first."
    )

X_train, y_train, _ = load_feature_matrix(FEATURES_TRAIN_PATH)
X_test,  y_test,  _ = load_feature_matrix(FEATURES_TEST_PATH)

print(f"X_train shape : {X_train.shape}")
print(f"X_test  shape : {X_test.shape}")
print()
print("Training class balance:")
for cls, cnt in zip(*np.unique(y_train, return_counts=True)):
    print(f"  Class {cls}: {cnt:>5d} samples ({cnt/len(y_train)*100:.1f}%)")
print()
print("Test class balance:")
for cls, cnt in zip(*np.unique(y_test, return_counts=True)):
    print(f"  Class {cls}: {cnt:>5d} samples ({cnt/len(y_test)*100:.1f}%)")
```

## Section 1 — Sub-question 1: Kernel & Block Normalisation Scheme

This section replicates **Sub-question 1**: finding the optimal SVM kernel function.

Features were extracted with **L1 block normalisation** and the default HOG parameters (pixels per cell = 8×8, cells per block = 3×3, orientation bins = 11). Multiple kernel/gamma configurations are compared via 5-fold cross-validation. `StandardScaler` is applied within each fold via the `Pipeline`, so normalisation is fitted only on training data — preventing leakage.

```python
# Note: The UndefinedMetricWarning below is expected for RBF configurations with extreme gamma
# values (γ = 1sd ≈ 0.025 through γ = 100). At these scales the RBF kernel becomes so localised
# that the decision boundary collapses and the SVM predicts only the majority class, leaving
# precision undefined. These configurations are correctly recorded as 0 in the results table.
kernel_results_df = run_kernel_grid_search(X_train, y_train, cv=CV_FOLDS)

print("Kernel Grid Search Results (sorted by mean F1, descending):\n")
display(kernel_results_df)
print("\nTop 5 Configurations:")
display(
    kernel_results_df.head(5).style.background_gradient(
        cmap="Greens", subset=["mean_accuracy", "mean_auc", "mean_f1"]
    )
)

fig, axes = plt.subplots(1, 3, figsize=(20, 8))
metrics = [
    ("mean_accuracy", "std_accuracy", "Accuracy"),
    ("mean_auc",      "std_auc",      "AUC"),
    ("mean_f1",       "std_f1",       "F1 Score"),
]
for ax, (mean_col, std_col, label) in zip(axes, metrics):
    ax.barh(kernel_results_df["kernel_label"], kernel_results_df[mean_col],
            xerr=kernel_results_df[std_col], color="steelblue", alpha=0.8, capsize=3)
    ax.set_title(f"Kernel Comparison — {label}", fontsize=13)
    ax.set_xlabel(label, fontsize=11)
    ax.set_ylabel("Kernel Configuration", fontsize=11)
    ax.set_xlim(0, 1)
    ax.grid(axis="x", linestyle="--", alpha=0.5)
plt.suptitle(
    "SVM Kernel Grid Search Results (5-fold CV, L1 HOG features) — Figures 12–14",
    fontsize=14, y=1.01
)
plt.tight_layout()
```

### Best Configuration Analysis

The top-ranked configuration by mean F1 is the **RBF kernel with γ = auto** (= 1/n_features).

**This kernel and γ value are fixed for Sub-questions 2 and 3.**

## Section 2 — Sub-question 2: Cell & Block Size

The best kernel from Section 1 (**RBF, γ = 1/n_features**) is fixed. The number of orientation bins is fixed at **11**, and block normalisation at **L1**.

We evaluate 12 combinations of `pixels_per_cell` ∈ {4, 8, 12, 16} and `cells_per_block` ∈ {2, 3, 4}. Larger cells capture coarser spatial gradient statistics; larger blocks provide more context for normalisation but reduce spatial resolution.

```python
# Note: The UndefinedMetricWarning below is expected for ppc=4 configurations with large block
# sizes (cpb=3 and cpb=4). At those settings the feature vector reaches 25,344–39,600 dimensions,
# making γ = 1/n_features extremely small; the SVM again predicts only the majority class,
# rendering precision undefined. Configurations unaffected by this collapse are scored normally.
GAMMA = 1 / X_train.shape[1]
print(f"Gamma (1 / n_features): {GAMMA:.6f}")

cell_block_combos = [
    ((4,  4), (2, 2)), ((4,  4), (3, 3)), ((4,  4), (4, 4)),
    ((8,  8), (2, 2)), ((8,  8), (3, 3)), ((8,  8), (4, 4)),
    ((12,12), (2, 2)), ((12,12), (3, 3)), ((12,12), (4, 4)),
    ((16,16), (2, 2)), ((16,16), (3, 3)), ((16,16), (4, 4)),
]
cell_block_configs = []
for ppc, cpb in cell_block_combos:
    path = make_ablation_npz_name(FEATURES_DIR, ppc, cpb, 11)
    if not os.path.exists(path):
        print(f"WARNING: Missing ablation file, skipping: {path}")
        continue
    cell_block_configs.append({
        "path": path, "label": f"ppc={ppc[0]},cpb={cpb[0]}",
        "pixels_per_cell": ppc, "cells_per_block": cpb, "orientations": 11,
    })
print(f"Loaded {len(cell_block_configs)} / {len(cell_block_combos)} cell/block configurations.")

cell_block_results_df = run_hog_ablation(cell_block_configs, kernel="rbf", gamma=GAMMA, cv=CV_FOLDS)
print("Cell/Block Ablation Results (sorted by mean F1, descending):\n")
display(cell_block_results_df)

# Plot: Accuracy & AUC side by side
fig, axes = plt.subplots(1, 2, figsize=(16, 7))
axes[0].barh(cell_block_results_df["label"], cell_block_results_df["mean_accuracy"],
             xerr=cell_block_results_df["std_accuracy"], color="steelblue", alpha=0.8, capsize=3)
axes[0].set_title("Cell/Block Size — Accuracy", fontsize=13)
axes[0].set_xlabel("Mean Accuracy (5-fold CV)", fontsize=11)
axes[0].set_ylabel("Configuration", fontsize=11)
axes[0].set_xlim(0, 1)
axes[0].grid(axis="x", linestyle="--", alpha=0.5)
axes[1].barh(cell_block_results_df["label"], cell_block_results_df["mean_auc"],
             xerr=cell_block_results_df["std_auc"], color="darkorange", alpha=0.8, capsize=3)
axes[1].set_title("Cell/Block Size — AUC", fontsize=13)
axes[1].set_xlabel("Mean AUC (5-fold CV)", fontsize=11)
axes[1].set_ylabel("Configuration", fontsize=11)
axes[1].set_xlim(0, 1)
axes[1].grid(axis="x", linestyle="--", alpha=0.5)
plt.suptitle(
    "HOG Cell/Block Size Ablation — Accuracy & AUC (RBF \u03b3=1/n_features, L1, ori=11) — Figure 15",
    fontsize=13
)
plt.tight_layout()

# Plot: F1
fig, ax = plt.subplots(figsize=(10, 7))
ax.barh(cell_block_results_df["label"], cell_block_results_df["mean_f1"],
        xerr=cell_block_results_df["std_f1"], color="seagreen", alpha=0.8, capsize=3)
ax.set_title("HOG Cell/Block Size Ablation — F1 Score — Figure 16", fontsize=13)
ax.set_xlabel("Mean F1 Score (5-fold CV)", fontsize=11)
ax.set_ylabel("Configuration", fontsize=11)
ax.set_xlim(0, 1)
ax.grid(axis="x", linestyle="--", alpha=0.5)
plt.tight_layout()

# Plot: Precision & Recall
fig, ax = plt.subplots(figsize=(12, 7))
y_pos = range(len(cell_block_results_df))
h = 0.35
ax.barh([p + h/2 for p in y_pos], cell_block_results_df["mean_precision"],
        height=h, xerr=cell_block_results_df["std_precision"],
        color="cornflowerblue", alpha=0.8, capsize=3, label="Precision")
ax.barh([p - h/2 for p in y_pos], cell_block_results_df["mean_recall"],
        height=h, xerr=cell_block_results_df["std_recall"],
        color="salmon", alpha=0.8, capsize=3, label="Recall")
ax.set_yticks(list(y_pos))
ax.set_yticklabels(cell_block_results_df["label"])
ax.set_title("HOG Cell/Block Size Ablation — Precision & Recall — Figure 17", fontsize=13)
ax.set_xlabel("Score (5-fold CV)", fontsize=11)
ax.set_ylabel("Configuration", fontsize=11)
ax.set_xlim(0, 1)
ax.legend()
ax.grid(axis="x", linestyle="--", alpha=0.5)
plt.tight_layout()
```

### Sub-question 2 Conclusion

The optimal cell/block configuration is the one ranked first in `cell_block_results_df` — i.e. the configuration achieving the **highest mean F1** with well-balanced precision and recall.

The results clearly show that **8×8-pixel cells** consistently outperform both finer (4×4) and coarser (12×12, 16×16) alternatives across all metrics. The top-ranked configuration by mean F1 is **ppc=8, cpb=4** (F1 = 0.628, AUC = 0.785), with **ppc=8, cpb=3** ranked closely behind (F1 = 0.604, AUC = 0.765). Configurations based on 12×12 cells fall noticeably below the 8×8 group, and 16×16 configurations deteriorate further (F1 ≤ 0.43), consistent with the loss of fine spatial gradient detail needed to distinguish panel edges from surrounding structures. At the opposite extreme, 4×4-pixel cells with large block sizes (cpb=3 or cpb=4) produce near-zero F1 scores: at these settings the feature vector reaches between 25,344 and 39,600 dimensions, and the correspondingly tiny γ = 1/n_features drives the RBF kernel into a regime where the SVM classifies virtually all samples as the majority class — the same mechanism that triggers the `UndefinedMetricWarning` observed during this cell's execution.

In practice this tends to be **ppc=8, cpb=3**: 8×8-pixel cells capture local gradient statistics at a scale matched to typical panel edge widths, while 3×3-cell blocks provide sufficient spatial context for contrast normalisation without over-smoothing.

**This cell/block size is fixed for Sub-question 3.**

## Section 3 — Sub-question 3: Orientation Bins

Cell/block size is fixed at the optimal from Section 2 (**ppc=8×8, cpb=3×3**). Block normalisation remains **L1**. The best kernel from Section 1 (**RBF, γ = 1/n_features**) is used.

We evaluate orientation counts ∈ {6, 8, 9, 11, 13, 18}. More bins capture finer gradient direction distinctions but increase feature dimensionality and can introduce noise in low-contrast regions.

```python
orientation_values = [6, 8, 9, 11, 13, 18]
orientation_configs = []
for ori in orientation_values:
    path = make_ablation_npz_name(FEATURES_DIR, (8, 8), (3, 3), ori)
    if not os.path.exists(path):
        print(f"WARNING: Missing ablation file, skipping: {path}")
        continue
    orientation_configs.append({
        "path": path, "label": f"ori={ori}",
        "pixels_per_cell": (8, 8), "cells_per_block": (3, 3), "orientations": ori,
    })
print(f"Loaded {len(orientation_configs)} / {len(orientation_values)} orientation configurations.")

orientation_results_df = run_hog_ablation(orientation_configs, kernel="rbf", gamma=GAMMA, cv=CV_FOLDS)
print("Orientation Ablation Results (sorted by mean F1, descending):\n")
display(orientation_results_df)

# Plot: Accuracy & AUC
fig, axes = plt.subplots(1, 2, figsize=(16, 6))
axes[0].barh(orientation_results_df["label"], orientation_results_df["mean_accuracy"],
             xerr=orientation_results_df["std_accuracy"], color="steelblue", alpha=0.8, capsize=3)
axes[0].set_title("Orientation Bins — Accuracy", fontsize=13)
axes[0].set_xlabel("Mean Accuracy (5-fold CV)", fontsize=11)
axes[0].set_ylabel("Orientation Bins", fontsize=11)
axes[0].set_xlim(0, 1)
axes[0].grid(axis="x", linestyle="--", alpha=0.5)
axes[1].barh(orientation_results_df["label"], orientation_results_df["mean_auc"],
             xerr=orientation_results_df["std_auc"], color="darkorange", alpha=0.8, capsize=3)
axes[1].set_title("Orientation Bins — AUC", fontsize=13)
axes[1].set_xlabel("Mean AUC (5-fold CV)", fontsize=11)
axes[1].set_ylabel("Orientation Bins", fontsize=11)
axes[1].set_xlim(0, 1)
axes[1].grid(axis="x", linestyle="--", alpha=0.5)
plt.suptitle(
    "HOG Orientation Bins Ablation — Accuracy & AUC (RBF \u03b3=1/n_features, ppc=8, cpb=3, L1) — Figure 18",
    fontsize=13
)
plt.tight_layout()

# Plot: F1
fig, ax = plt.subplots(figsize=(9, 6))
ax.barh(orientation_results_df["label"], orientation_results_df["mean_f1"],
        xerr=orientation_results_df["std_f1"], color="seagreen", alpha=0.8, capsize=3)
ax.set_title("HOG Orientation Bins Ablation — F1 Score — Figure 19", fontsize=13)
ax.set_xlabel("Mean F1 Score (5-fold CV)", fontsize=11)
ax.set_ylabel("Orientation Bins", fontsize=11)
ax.set_xlim(0, 1)
ax.grid(axis="x", linestyle="--", alpha=0.5)
plt.tight_layout()

# Plot: Precision & Recall
fig, ax = plt.subplots(figsize=(11, 6))
y_pos = range(len(orientation_results_df))
h = 0.35
ax.barh([p + h/2 for p in y_pos], orientation_results_df["mean_precision"],
        height=h, xerr=orientation_results_df["std_precision"],
        color="cornflowerblue", alpha=0.8, capsize=3, label="Precision")
ax.barh([p - h/2 for p in y_pos], orientation_results_df["mean_recall"],
        height=h, xerr=orientation_results_df["std_recall"],
        color="salmon", alpha=0.8, capsize=3, label="Recall")
ax.set_yticks(list(y_pos))
ax.set_yticklabels(orientation_results_df["label"])
ax.set_title("HOG Orientation Bins Ablation — Precision & Recall — Figure 20", fontsize=13)
ax.set_xlabel("Score (5-fold CV)", fontsize=11)
ax.set_ylabel("Orientation Bins", fontsize=11)
ax.set_xlim(0, 1)
ax.legend()
ax.grid(axis="x", linestyle="--", alpha=0.5)
plt.tight_layout()
```

### Sub-question 3 Conclusion

The optimal orientation count is the one ranked first in `orientation_results_df` — the configuration with the **highest mean F1**.

This is consistent with the rectangular geometry of solar panels: panels produce strong, directionally consistent gradient responses, and 11 bins is fine enough to distinguish panel edges from surrounding vegetation or rooftops without capturing spurious noise directions.

Combined with the findings from Sections 1 and 2, the optimal HOG+SVM configuration is now **fully identified**: RBF kernel, γ = 1/n_features, ppc=8×8, cpb=3×3, 11 orientation bins, L1 block normalisation.

The orientation-bin ablation reveals a remarkably flat performance landscape: all six configurations achieve mean F1 scores within the narrow range 0.587–0.604 and mean AUC values between 0.762 and 0.767. **ori=11** ranks first with F1 = 0.604 and AUC = 0.765, but the margin over ori=8 (F1 = 0.603, AUC = 0.767) is negligible. A modest downward trend in F1 is visible as bins increase beyond 11 (F1 = 0.596 at ori=13, 0.587 at ori=18), suggesting that very fine angular resolution introduces redundant gradient direction distinctions without adding discriminative power. Reducing bins below 11 (to 6 or 8) degrades recall slightly while holding precision relatively stable. These small effect sizes indicate that orientation count is a comparatively insensitive hyperparameter for this task, and that the dominant panel-edge gradients are captured adequately across the full tested range. The choice of **11 bins** represents a defensible optimum: it encodes sufficient directional variation while remaining robust to noise, and is consistent with the recommendation adopted in the thesis.

## Section 4 — Train Final Model

The three sub-questions converge on the following optimal configuration:

| Parameter | Value |
|-----------|-------|
| Kernel | RBF |
| γ (gamma) | 1 / n_features |
| C | 1.0 |
| HOG block normalisation | L1 |
| Orientation bins | 11 |
| Pixels per cell | 8×8 |
| Cells per block | 3×3 |

The final model is trained on the **full training set** (no held-out validation fold). Generalisation performance is assessed on the independent test set in Section 5.

```python
final_model = train_svm(X_train, y_train, kernel="rbf", C=1.0, gamma=GAMMA, probability=True)
train_acc = final_model.score(X_train, y_train)
print(f"Training accuracy (sanity check): {train_acc:.4f}")
```

## Section 5 — Holdout Evaluation

The final model is evaluated on the **held-out test set**: samples that were never seen during training or any cross-validation fold. This provides an unbiased estimate of real-world generalisation performance.

```python
holdout_results = evaluate_on_holdout(final_model, X_test, y_test)

print("Holdout Test Set Results")
print("=" * 35)
for metric, label in [
    ("accuracy",  "Accuracy "),
    ("precision", "Precision"),
    ("recall",    "Recall   "),
    ("f1",        "F1 Score "),
    ("roc_auc",   "AUC      "),
]:
    print(f"  {label} : {holdout_results[metric]:.4f}")

fig, axes = plt.subplots(1, 3, figsize=(18, 5))
plot_confusion_matrix(holdout_results["confusion_matrix"], title="Confusion Matrix", ax=axes[0])
plot_roc_curve(y_test, holdout_results["y_prob"], title="ROC Curve", ax=axes[1])
plot_precision_recall_curve(y_test, holdout_results["y_prob"], title="Precision-Recall Curve", ax=axes[2])
plt.suptitle("Final Model — Holdout Evaluation", fontsize=14)
plt.tight_layout()
```

### Interpretation

The final HOG+SVM model achieves **75.4% accuracy** and an **AUC of 0.823** on the held-out test set. Holdout scores exceed their 5-fold CV counterparts across every metric — accuracy (+4.5 pp), AUC (+5.8 pp), F1 (+5.4 pp), precision (+3.2 pp), and recall (+6.5 pp) — indicating that the model generalises well and does not overfit. This positive CV-to-holdout gap is consistent with the 4× rotational augmentation applied during preprocessing: training on images at all four cardinal orientations builds broad robustness, while the held-out test set (drawn from the original non-augmented distribution) represents a comparatively contained challenge. **Precision (0.71) exceeds recall (0.61)**, meaning the classifier is more conservative about predicting panel presence — when it fires, it is reasonably reliable, but it misses roughly one in three positive tiles. The **ROC curve** (AUC = 0.823) confirms a strong ability to rank positive tiles above negatives across all decision thresholds; the **precision-recall curve** reflects the moderate recall limitation while showing that precision holds up well at the default 0.5 operating point. Overall, these results validate the chosen HOG+SVM configuration as a competent and reproducible baseline for aerial solar panel detection.

## Section 6 — Results Summary

Consolidated table comparing 5-fold CV performance (on the training set, using the default HOG configuration) with final holdout test set performance.

```python
rbf_row = kernel_results_df[kernel_results_df["kernel_label"] == "RBF \u03b3=auto"].iloc[0]
summary_df = pd.DataFrame({
    "Metric": ["Accuracy", "AUC", "F1 Score", "Precision", "Recall"],
    "CV Mean": [
        rbf_row["mean_accuracy"], rbf_row["mean_auc"], rbf_row["mean_f1"],
        rbf_row["mean_precision"], rbf_row["mean_recall"],
    ],
    "CV Std": [
        rbf_row["std_accuracy"],  rbf_row["std_auc"],  rbf_row["std_f1"],
        rbf_row["std_precision"],  rbf_row["std_recall"],
    ],
    "Holdout Score": [
        holdout_results["accuracy"], holdout_results["roc_auc"],
        holdout_results["f1"], holdout_results["precision"],
        holdout_results["recall"],
    ],
})
summary_df["CV Mean"]       = summary_df["CV Mean"].map("{:.4f}".format)
summary_df["CV Std"]        = summary_df["CV Std"].map("{:.4f}".format)
summary_df["Holdout Score"] = summary_df["Holdout Score"].map("{:.4f}".format)
display(summary_df.set_index("Metric"))
```

### Suggested next steps:

- **More data:** acquiring additional labelled tiles from the DeepSolaris or DeepGeoStat dataset would likely improve generalisation further.
- **Higher resolution imagery:** current tiles are resized to a fixed resolution; higher-resolution input could reveal finer texture and edge cues.
- **Deep learning comparison:** a CNN (e.g. ResNet-18) or Vision Transformer fine-tuned on this dataset would provide a natural comparison point. HOG+SVM is interpretable and fast to train, but deep features may capture non-linear panel appearance patterns more effectively.
- **Confidence calibration:** probability calibration (Platt scaling or isotonic regression) would make the predicted probabilities more reliable for downstream thresholding.
