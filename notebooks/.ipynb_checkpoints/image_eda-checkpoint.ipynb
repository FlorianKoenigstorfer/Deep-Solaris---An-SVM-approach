## Section 0 — Setup & Imports

```python
%matplotlib inline

import os
import sys

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from PIL import Image

sys.path.insert(0, "../src")

from image_eda import (
    compute_mean_image,
    compute_pixel_stats,
    detect_duplicates,
    load_image_manifest,
)

sns.set_theme(style="whitegrid")
RANDOM_SEED = 42
DATA_ROOT = "../data"

expected_dirs = [
    os.path.join(DATA_ROOT, "Train", "Positives"),
    os.path.join(DATA_ROOT, "Train", "Negatives"),
    os.path.join(DATA_ROOT, "Test", "Positives"),
    os.path.join(DATA_ROOT, "Test", "Negatives"),
]
missing_dirs = [p for p in expected_dirs if not os.path.isdir(p)]
if missing_dirs:
    print("Warning: Some expected dataset folders are missing:")
    for p in missing_dirs:
        print(f"  - {p}")

manifest_df = load_image_manifest(DATA_ROOT)
required_manifest_cols = {"filepath", "split", "label", "is_corrupt", "width", "height", "mode", "file_size_bytes"}
missing_manifest_cols = required_manifest_cols - set(manifest_df.columns)
if missing_manifest_cols:
    raise ValueError(f"Manifest is missing required columns: {sorted(missing_manifest_cols)}")

pixel_stats = compute_pixel_stats(manifest_df)
eda_df = manifest_df.merge(pixel_stats, on="filepath", how="left")

# Backward-compatible alias used by later cells.
df = eda_df.copy()

print(f"Loaded manifest rows: {len(eda_df)}")
print(f"Corrupt rows: {int(eda_df['is_corrupt'].sum())}")
print(f"Columns available: {sorted(eda_df.columns.tolist())}")
```

## Section 1 — Dataset Overview

```python
summary = {
    "total_images": len(df),
    "corrupt_images": int(df["is_corrupt"].sum()),
}
summary_table = pd.DataFrame([summary])
print("Summary table:")
display(summary_table)

counts = (
    df.groupby(["split", "label"], dropna=False)
    .size()
    .reset_index(name="count")
    .sort_values(["split", "label"])
)
print("Counts per split x label:")
display(counts)

print("Class balance ratio (positives / negatives) by split:")
for split_name in ["train", "test"]:
    split_df = df[df["split"] == split_name]
    pos = (split_df["label"] == "positive").sum()
    neg = (split_df["label"] == "negative").sum()
    ratio = np.nan if neg == 0 else pos / neg
    print(f"  {split_name}: {ratio}")

plt.figure(figsize=(8, 5))
ax = sns.barplot(data=counts, x="split", y="count", hue="label", palette="Greys")
ax.set_title("Class Distribution by Split")
ax.set_xlabel("Split")
ax.set_ylabel("Image Count")
ax.legend(title="Label")
```

## Section 2 — Sample Images

```python
train_clean = df[(df["split"] == "train") & (~df["is_corrupt"])].copy()

fig, axes = plt.subplots(2, 5, figsize=(18, 7))
for row_idx, label in enumerate(["positive", "negative"]):
    label_files = train_clean[train_clean["label"] == label]["filepath"]
    sample_n = min(5, len(label_files))
    sampled = label_files.sample(n=sample_n, random_state=RANDOM_SEED) if sample_n > 0 else pd.Series(dtype=object)

    for col_idx in range(5):
        ax = axes[row_idx, col_idx]
        ax.set_xticks([])
        ax.set_yticks([])
        if col_idx < len(sampled):
            fp = sampled.iloc[col_idx]
            try:
                with Image.open(fp) as img:
                    ax.imshow(img.convert("RGB"))
                ax.set_title(os.path.basename(fp)[:20], fontsize=9)
            except Exception:
                ax.text(0.5, 0.5, "Load error", ha="center", va="center")
        else:
            ax.text(0.5, 0.5, "N/A", ha="center", va="center")

axes[0, 0].set_ylabel("Positives", fontsize=11)
axes[1, 0].set_ylabel("Negatives", fontsize=11)
fig.suptitle("Random Training Samples by Class")
```

## Section 3 — Image Dimensions & Consistency

```python
dim_df = df[~df["is_corrupt"]].copy()
size_counts = (
    dim_df.groupby(["width", "height"]).size().reset_index(name="count").sort_values("count", ascending=False)
)
print("Unique width x height combinations:")
display(size_counts)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
sns.histplot(data=dim_df, x="width", hue="label", alpha=0.5, ax=axes[0], element="step")
axes[0].set_title("Width Distribution by Label")
axes[0].set_xlabel("Width")
axes[0].set_ylabel("Count")

sns.histplot(data=dim_df, x="height", hue="label", alpha=0.5, ax=axes[1], element="step")
axes[1].set_title("Height Distribution by Label")
axes[1].set_xlabel("Height")
axes[1].set_ylabel("Count")

plt.figure(figsize=(7, 6))
ax = sns.scatterplot(data=dim_df, x="width", y="height", hue="label", alpha=0.6)
ax.set_title("Width vs Height by Label")
ax.set_xlabel("Width")
ax.set_ylabel("Height")
ax.legend(title="Label")

dim_df["aspect_ratio"] = dim_df["width"] / dim_df["height"]
plt.figure(figsize=(8, 5))
ax = sns.histplot(data=dim_df, x="aspect_ratio", hue="label", alpha=0.5, kde=True, element="step")
ax.set_title("Aspect Ratio Distribution by Label")
ax.set_xlabel("Aspect Ratio (width / height)")
ax.set_ylabel("Count")

all_same_size = dim_df[["width", "height"]].nunique().eq(1).all()
print(f"all_same_size = {all_same_size}")
if all_same_size:
    print("All non-corrupt images have identical dimensions.")
else:
    print("Image dimensions vary across the dataset.")
```

## Section 4 — Color Space Analysis

```python
section4_df = eda_df.copy() if "eda_df" in globals() else df.copy()
core_cols = {"filepath", "split", "label", "is_corrupt", "mode"}
missing_core_cols = core_cols - set(section4_df.columns)

# Attempt self-recovery when the working DataFrame has been modified.
if missing_core_cols:
    print(f"Section 4 detected missing columns {sorted(missing_core_cols)}. Rebuilding from DATA_ROOT...")
    rebuilt_manifest = load_image_manifest(DATA_ROOT)
    rebuilt_pixel_stats = compute_pixel_stats(rebuilt_manifest)
    section4_df = rebuilt_manifest.merge(rebuilt_pixel_stats, on="filepath", how="left")

missing_core_cols = core_cols - set(section4_df.columns)
if missing_core_cols:
    raise ValueError(f"Section 4 cannot run. Missing required columns: {sorted(missing_core_cols)}")

valid_df = section4_df[~section4_df["is_corrupt"]].copy()

mode_counts = (
    valid_df.groupby(["split", "label", "mode"])
    .size()
    .reset_index(name="count")
    .sort_values(["split", "label", "count"], ascending=[True, True, False])
)
print("Mode counts by split x label:")
display(mode_counts)

non_rgb_count = int((valid_df["mode"] != "RGB").sum())
if non_rgb_count > 0:
    print(f"Found {non_rgb_count} non-RGB images.")
else:
    print("All non-corrupt images are RGB.")

stats_cols = ["mean_r", "mean_g", "mean_b", "std_r", "std_g", "std_b", "mean_brightness"]
missing_stats_cols = [col for col in stats_cols if col not in section4_df.columns]
if missing_stats_cols:
    print(f"Section 4 is missing stats columns {missing_stats_cols}. Recomputing pixel stats...")
    refreshed_stats = compute_pixel_stats(section4_df)
    section4_df = section4_df.drop(columns=[c for c in refreshed_stats.columns if c != "filepath"], errors="ignore")
    section4_df = section4_df.merge(refreshed_stats, on="filepath", how="left")

train_stats = section4_df[(section4_df["split"] == "train") & (~section4_df["is_corrupt"])].dropna(
    subset=["mean_r", "mean_g", "mean_b"]
)

if train_stats.empty:
    print("No non-corrupt training images with pixel stats are available for Section 4 plots.")

agg = (
    train_stats.groupby("label")[["mean_r", "mean_g", "mean_b", "std_r", "std_g", "std_b"]]
    .mean()
    .reset_index()
    if not train_stats.empty
    else pd.DataFrame(columns=["label", "mean_r", "mean_g", "mean_b", "std_r", "std_g", "std_b"])
)

plot_rows = []
for _, row in agg.iterrows():
    plot_rows.extend(
        [
            {"label": row["label"], "channel": "R", "mean": row["mean_r"], "err": row["std_r"]},
            {"label": row["label"], "channel": "G", "mean": row["mean_g"], "err": row["std_g"]},
            {"label": row["label"], "channel": "B", "mean": row["mean_b"], "err": row["std_b"]},
        ]
    )
channel_plot_df = pd.DataFrame(plot_rows, columns=["label", "channel", "mean", "err"])

plt.figure(figsize=(9, 5))
if channel_plot_df.empty:
    plt.title("Mean Channel Values by Label (Training Set)")
    plt.xlabel("Channel")
    plt.ylabel("Mean Pixel Value")
    plt.text(0.5, 0.5, "No data available", ha="center", va="center")
else:
    ax = sns.barplot(data=channel_plot_df, x="channel", y="mean", hue="label", palette="muted", errorbar=None)
    for idx, patch in enumerate(ax.patches):
        err = channel_plot_df.iloc[idx]["err"] if idx < len(channel_plot_df) else 0
        x = patch.get_x() + patch.get_width() / 2
        y = patch.get_height()
        ax.errorbar(x=x, y=y, yerr=err, fmt="none", c="black", capsize=3)
    ax.set_title("Mean Channel Values by Label (Training Set)")
    ax.set_xlabel("Channel")
    ax.set_ylabel("Mean Pixel Value")
    ax.legend(title="Label")

sample_200 = (
    train_stats.groupby("label", group_keys=False)
    .apply(lambda g: g.sample(n=min(200, len(g)), random_state=RANDOM_SEED))
    .reset_index(drop=True)
    if not train_stats.empty
    else pd.DataFrame(columns=train_stats.columns)
)
plt.figure(figsize=(8, 5))
if sample_200.empty:
    plt.title("Brightness Distribution by Label")
    plt.xlabel("Mean Brightness")
    plt.ylabel("Density")
    plt.text(0.5, 0.5, "No data available", ha="center", va="center")
else:
    for label, group in sample_200.groupby("label"):
        sns.kdeplot(data=group, x="mean_brightness", fill=False, label=label)
    plt.title("Brightness Distribution by Label")
    plt.xlabel("Mean Brightness")
    plt.ylabel("Density")
    plt.legend(title="Label")

pos_paths = train_stats[train_stats["label"] == "positive"]["filepath"].tolist() if "label" in train_stats.columns else []
neg_paths = train_stats[train_stats["label"] == "negative"]["filepath"].tolist() if "label" in train_stats.columns else []

pos_mean_img = compute_mean_image(pos_paths, target_size=(128, 128))
neg_mean_img = compute_mean_image(neg_paths, target_size=(128, 128))

fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(pos_mean_img)
axes[0].set_title("Mean Image - Positives (Train)")
axes[0].set_xlabel("X")
axes[0].set_ylabel("Y")
axes[0].set_xticks([])
axes[0].set_yticks([])

axes[1].imshow(neg_mean_img)
axes[1].set_title("Mean Image - Negatives (Train)")
axes[1].set_xlabel("X")
axes[1].set_ylabel("Y")
axes[1].set_xticks([])
axes[1].set_yticks([])
```

## Section 5 — Brightness & Contrast

```python
stats_df = df[(~df["is_corrupt"]) & df["mean_brightness"].notna()].copy()

fig, axes = plt.subplots(1, 2, figsize=(14, 5), sharey=True)
for idx, split_name in enumerate(["train", "test"]):
    sub = stats_df[stats_df["split"] == split_name]
    for label, g in sub.groupby("label"):
        sns.kdeplot(data=g, x="mean_brightness", fill=True, alpha=0.4, label=label, ax=axes[idx])
    axes[idx].set_title(f"Brightness Distribution: {split_name.capitalize()}")
    axes[idx].set_xlabel("Mean Brightness")
    axes[idx].set_ylabel("Density")
    axes[idx].legend(title="Label")

fig.suptitle("Brightness Distribution: Train vs Test")

plt.figure(figsize=(8, 5))
for label, g in stats_df.groupby("label"):
    sns.kdeplot(data=g, x="std_brightness", fill=True, alpha=0.4, label=label)
plt.title("Contrast Distribution by Label")
plt.xlabel("Brightness Std Dev")
plt.ylabel("Density")
plt.legend(title="Label")

very_dark = stats_df[stats_df["mean_brightness"] < 20]
very_bright = stats_df[stats_df["mean_brightness"] > 235]

print("Very dark image counts by split x label:")
display(very_dark.groupby(["split", "label"]).size().reset_index(name="count"))
print("Very bright image counts by split x label:")
display(very_bright.groupby(["split", "label"]).size().reset_index(name="count"))

for title, subset in [("Very Dark Examples", very_dark), ("Very Bright Examples", very_bright)]:
    examples = subset.sample(n=min(3, len(subset)), random_state=RANDOM_SEED) if len(subset) > 0 else pd.DataFrame()
    fig, axes = plt.subplots(1, 3, figsize=(12, 4))
    fig.suptitle(title)
    for i in range(3):
        ax = axes[i]
        ax.set_xticks([])
        ax.set_yticks([])
        if i < len(examples):
            fp = examples.iloc[i]["filepath"]
            try:
                with Image.open(fp) as img:
                    ax.imshow(img.convert("RGB"))
                ax.set_title(os.path.basename(fp)[:20])
            except Exception:
                ax.text(0.5, 0.5, "Load error", ha="center", va="center")
        else:
            ax.text(0.5, 0.5, "N/A", ha="center", va="center")
```

## Section 6 — Data Quality Issues

```python
corrupt_paths = df[df["is_corrupt"]]["filepath"].tolist()
print("Corrupt file paths:")
if corrupt_paths:
    for p in corrupt_paths:
        print(p)
else:
    print("None")

df_dups = detect_duplicates(df)
duplicate_hashes = df_dups[df_dups["is_duplicate"]]["file_hash"].dropna().nunique()
images_involved = int(df_dups["is_duplicate"].sum())

print(f"Duplicate hashes found: {duplicate_hashes}")
print(f"Images involved in duplicates: {images_involved}")

if images_involved > 0:
    dup_summary = (
        df_dups[df_dups["is_duplicate"]]
        .groupby(["file_hash", "split", "label"])
        .size()
        .reset_index(name="count")
        .sort_values(["file_hash", "split", "label"])
    )
    print("Duplicate groups by split/label:")
    display(dup_summary)

mixed_mode_count = int(((df_dups["mode"] != "RGB") & (~df_dups["is_corrupt"])).sum())
extreme_brightness_count = int(((df_dups["mean_brightness"] < 20) | (df_dups["mean_brightness"] > 235)).sum())

quality_summary = pd.DataFrame(
    [
        {
            "total_images": len(df_dups),
            "corrupt_count": int(df_dups["is_corrupt"].sum()),
            "duplicate_count": images_involved,
            "mixed_mode_count": mixed_mode_count,
            "extreme_brightness_count": extreme_brightness_count,
        }
    ]
)
print("Quality summary table:")
display(quality_summary)
```

## Section 7 — Train / Test Consistency Check

```python
metrics = ["mean_brightness", "std_brightness", "mean_r", "mean_g", "mean_b"]
consistency_df = (eda_df.copy() if "eda_df" in globals() else df.copy())

fig, axes = plt.subplots(3, 2, figsize=(14, 14))
axes = axes.flatten()

for i, metric in enumerate(metrics):
    ax = axes[i]
    plotted_any = False
    for label in ["positive", "negative"]:
        for split_name, line_style in [("train", "-"), ("test", "--")]:
            subset = consistency_df[
                (consistency_df["label"] == label)
                & (consistency_df["split"] == split_name)
                & consistency_df[metric].notna()
            ]
            if len(subset) > 1:
                plotted_any = True
                sns.kdeplot(
                    data=subset,
                    x=metric,
                    ax=ax,
                    label=f"{label}-{split_name}",
                    linestyle=line_style,
                    fill=False,
                )
    ax.set_title(f"{metric}: Train vs Test by Label")
    ax.set_xlabel(metric)
    ax.set_ylabel("Density")
    if plotted_any:
        ax.legend(title="Group", fontsize=8)

axes[-1].axis("off")
fig.suptitle("Train/Test Consistency Distributions", y=1.02)

metric_summary = (
    consistency_df.groupby(["split", "label"])[metrics]
    .agg(["mean", "std"])
    .round(3)
)
print("Mean and std of metrics per split and label:")
display(metric_summary)
```

A large discrepancy between train and test distributions (for brightness, contrast, or color channels) suggests dataset shift. This can reduce generalization, cause unstable validation performance, and lead the model to learn split-specific artifacts instead of class-relevant patterns.
