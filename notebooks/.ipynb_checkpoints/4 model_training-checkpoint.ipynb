# Solar Panel SVM Classifier — Model Training & Evaluation

**Goal:** Systematically investigate three sub-questions to identify the optimal
HOG+SVM configuration for solar panel detection, train the final model, and
evaluate on the holdout test set.

**Data source:** `.npz` feature files in `/data/` produced by `feature_extraction.ipynb`.

> *Note:* The findings from this analysis informed the work presented in
> Koenigstorfer, J. (2018). *Solar panel detection from aerial imagery using
> HOG features and Support Vector Machines*. MSc thesis, Maastricht University.

## Introduction

This notebook trains and evaluates a Support Vector Machine (SVM) classifier on
HOG (Histogram of Oriented Gradients) features extracted from aerial imagery.
Three sub-questions are systematically investigated to identify the optimal
configuration:

1. **Sub-question 1:** Which SVM kernel function is optimal?
2. **Sub-question 2:** Which HOG cell/block size combination is optimal?
3. **Sub-question 3:** Which number of HOG orientation bins is optimal?

All cross-validation uses a `Pipeline(StandardScaler, SVC)` so the scaler is
fitted only on training folds, preventing data leakage.

## Setup & Imports

```python
%matplotlib inline
import os
import sys
import warnings

# ── Warning suppression ───────────────────────────────────────────────────────
os.environ["PYTHONWARNINGS"] = "ignore::FutureWarning,ignore::UserWarning"

from sklearn.exceptions import ConvergenceWarning
warnings.filterwarnings("ignore", category=ConvergenceWarning)
warnings.filterwarnings("ignore", category=FutureWarning)
warnings.filterwarnings("ignore", category=UserWarning, module="pandas")

# ── Monkey-patch for CPython 3.13 ResourceTracker bug (cpython#140485) ────────
# The multiprocessing resource_tracker raises a harmless ChildProcessError
# during __del__ cleanup when used with joblib's loky backend on
# Python >= 3.12.10 / 3.13.x. Fixed in joblib >= 1.5.0 (loky >= 3.5.5).
# Remove this patch after upgrading joblib.
if sys.version_info >= (3, 12):
    import multiprocessing.resource_tracker as _mrt
    _OriginalRT = _mrt.ResourceTracker
    class _PatchedRT(_OriginalRT):
        def __del__(self):
            try:
                super().__del__()
            except ChildProcessError:
                pass
    _mrt.ResourceTracker = _PatchedRT
    # Replace the module-level singleton so all future users (including
    # joblib/loky, which has NOT been imported yet) get the patched version.
    _mrt._resource_tracker = _PatchedRT()
    _mrt.ensure_running = _mrt._resource_tracker.ensure_running
    _mrt.register = _mrt._resource_tracker.register
    _mrt.maybe_unlink = _mrt._resource_tracker.maybe_unlink
    _mrt.unregister = _mrt._resource_tracker.unregister
    _mrt.getfd = _mrt._resource_tracker.getfd
    del _mrt, _OriginalRT, _PatchedRT

# ── Imports (joblib/sklearn/model are loaded AFTER the patch) ─────────────────
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
sys.path.insert(0, "../src")
from model import (
    train_svm,
    cross_validate_svm,
    evaluate_on_holdout,
    plot_roc_curve,
    plot_precision_recall_curve,
    plot_confusion_matrix,
    plot_learning_curve,
    run_kernel_grid_search,
    run_hog_ablation,
)
from feature_extraction import load_feature_matrix, make_ablation_npz_name

# ── Constants ──────────────────────────────────────────────────────────────────
FEATURES_DIR        = "../data"
FEATURES_TRAIN_PATH = "../data/features_train.npz"
FEATURES_TEST_PATH  = "../data/features_test.npz"
CV_FOLDS            = 5
RANDOM_STATE        = 42
BLOCK_NORM          = "L1"


if not os.path.exists(FEATURES_TRAIN_PATH):
    raise FileNotFoundError(
        f"Training features not found at '{FEATURES_TRAIN_PATH}'. "
        "Please run feature_extraction.ipynb first."
    )
if not os.path.exists(FEATURES_TEST_PATH):
    raise FileNotFoundError(
        f"Test features not found at '{FEATURES_TEST_PATH}'. "
        "Please run feature_extraction.ipynb first."
    )

X_train, y_train, _ = load_feature_matrix(FEATURES_TRAIN_PATH)
X_test,  y_test,  _ = load_feature_matrix(FEATURES_TEST_PATH)

print(f"X_train shape : {X_train.shape}")
print(f"X_test  shape : {X_test.shape}")
print()
print("Training class balance:")
for cls, cnt in zip(*np.unique(y_train, return_counts=True)):
    print(f"  Class {cls}: {cnt:>5d} samples ({cnt / len(y_train) * 100:.1f}%)")
print()
print("Test class balance:")
for cls, cnt in zip(*np.unique(y_test, return_counts=True)):
    print(f"  Class {cls}: {cnt:>5d} samples ({cnt / len(y_test) * 100:.1f}%)")
```

## Section 1 — Sub-question 1: Kernel & Block Normalization Scheme

Replicates Sub-question 1 from the thesis — finding the optimal SVM kernel
function. Features were extracted with **L1 block normalization** (the fixed
optimal scheme). This section tests multiple kernel/gamma configurations via
5-fold cross-validation.

`StandardScaler` is applied within each CV fold via a `Pipeline` to prevent
data leakage — no scaling information from validation folds leaks into training.

```python
kernel_results_df = run_kernel_grid_search(X_train, y_train, cv=CV_FOLDS)

print("Kernel Grid Search Results (sorted by mean F1, descending):\n")
display(kernel_results_df)
print("\nTop 5 Configurations:")
display(
    kernel_results_df.head(5).style.background_gradient(
        cmap="Greens", subset=["mean_accuracy", "mean_auc", "mean_f1"]
    )
)

fig, axes = plt.subplots(1, 3, figsize=(20, 8))

metrics = [
    ("mean_accuracy", "std_accuracy", "Accuracy"),
    ("mean_auc",      "std_auc",      "AUC"),
    ("mean_f1",       "std_f1",       "F1 Score"),
]

for ax, (mean_col, std_col, label) in zip(axes, metrics):
    ax.barh(
        kernel_results_df["kernel_label"],
        kernel_results_df[mean_col],
        xerr=kernel_results_df[std_col],
        color="steelblue",
        alpha=0.8,
        capsize=3,
    )
    ax.set_title(f"Kernel Comparison — {label}", fontsize=13)
    ax.set_xlabel(label, fontsize=11)
    ax.set_ylabel("Kernel Configuration", fontsize=11)
    ax.set_xlim(0, 1)
    ax.grid(axis="x", linestyle="--", alpha=0.5)

plt.suptitle(
    "SVM Kernel Grid Search Results (5-fold CV, L1 HOG features) — Figures 12–14",
    fontsize=14, y=1.01,
)
plt.tight_layout()
```

### Best Configuration Analysis

The bar charts above show cross-validated accuracy, AUC, and F1 for each
kernel/gamma combination. The top-ranked configuration by mean F1 is
**RBF with `gamma='auto'`**, which in scikit-learn sets γ = 1 / n_features.
For this dataset, that corresponds to γ ≈ 0.000206. The accuracy and AUC bar
charts confirm that this configuration also leads on those metrics. This
configuration is fixed for Sub-questions 2 and 3.

## Section 2 — Sub-question 2: Cell & Block Size

Tests which HOG cell/block size combination is optimal. The best kernel from
Section 1 (**RBF, γ = 1 / n_features**) is fixed. Orientations are fixed at
**11**, block normalisation at **L1**.

Each configuration corresponds to a pre-saved `.npz` ablation file produced by
`feature_extraction.ipynb`. Files are loaded and evaluated with 5-fold CV.

```python
GAMMA = 1 / X_train.shape[1]
print(f"Gamma (1 / n_features): {GAMMA:.6f}")
cell_block_combos = [
    ((4,  4),  (2, 2)), ((4,  4),  (3, 3)), ((4,  4),  (4, 4)),
    ((8,  8),  (2, 2)), ((8,  8),  (3, 3)), ((8,  8),  (4, 4)),
    ((12, 12), (2, 2)), ((12, 12), (3, 3)), ((12, 12), (4, 4)),
    ((16, 16), (2, 2)), ((16, 16), (3, 3)), ((16, 16), (4, 4)),
]

cell_block_configs = []
for ppc, cpb in cell_block_combos:
    path = make_ablation_npz_name(FEATURES_DIR, ppc, cpb, 11)
    if not os.path.exists(path):
        print(f"WARNING: Missing ablation file, skipping: {path}")
        continue
    cell_block_configs.append({
        "path":            path,
        "label":           f"ppc={ppc[0]},cpb={cpb[0]}",
        "pixels_per_cell": ppc,
        "cells_per_block": cpb,
        "orientations":    11,
    })

print(f"Loaded {len(cell_block_configs)} / {len(cell_block_combos)} cell/block configurations.")

cell_block_results_df = run_hog_ablation(
    cell_block_configs, kernel="rbf", gamma=GAMMA, cv=CV_FOLDS
)
print("Cell/Block Ablation Results (sorted by mean F1, descending):\n")
display(cell_block_results_df)

fig, axes = plt.subplots(1, 2, figsize=(16, 7))

axes[0].barh(
    cell_block_results_df["label"], cell_block_results_df["mean_accuracy"],
    xerr=cell_block_results_df["std_accuracy"], color="steelblue", alpha=0.8, capsize=3,
)
axes[0].set_title("Cell/Block Size — Accuracy", fontsize=13)
axes[0].set_xlabel("Mean Accuracy (5-fold CV)", fontsize=11)
axes[0].set_ylabel("Configuration", fontsize=11)
axes[0].set_xlim(0, 1)
axes[0].grid(axis="x", linestyle="--", alpha=0.5)

axes[1].barh(
    cell_block_results_df["label"], cell_block_results_df["mean_auc"],
    xerr=cell_block_results_df["std_auc"], color="darkorange", alpha=0.8, capsize=3,
)
axes[1].set_title("Cell/Block Size — AUC", fontsize=13)
axes[1].set_xlabel("Mean AUC (5-fold CV)", fontsize=11)
axes[1].set_ylabel("Configuration", fontsize=11)
axes[1].set_xlim(0, 1)
axes[1].grid(axis="x", linestyle="--", alpha=0.5)

plt.suptitle(
    "HOG Cell/Block Size Ablation — Accuracy & AUC (RBF γ=1/n_features, L1, ori=11) — Figure 15",
    fontsize=13,
)
plt.tight_layout()

fig, ax = plt.subplots(figsize=(10, 7))
ax.barh(
    cell_block_results_df["label"], cell_block_results_df["mean_f1"],
    xerr=cell_block_results_df["std_f1"], color="seagreen", alpha=0.8, capsize=3,
)
ax.set_title("HOG Cell/Block Size Ablation — F1 Score — Figure 16", fontsize=13)
ax.set_xlabel("Mean F1 Score (5-fold CV)", fontsize=11)
ax.set_ylabel("Configuration", fontsize=11)
ax.set_xlim(0, 1)
ax.grid(axis="x", linestyle="--", alpha=0.5)
plt.tight_layout()

fig, ax = plt.subplots(figsize=(12, 7))
y_pos = range(len(cell_block_results_df))
h = 0.35

ax.barh(
    [p + h / 2 for p in y_pos], cell_block_results_df["mean_precision"],
    height=h, xerr=cell_block_results_df["std_precision"],
    color="cornflowerblue", alpha=0.8, capsize=3, label="Precision",
)
ax.barh(
    [p - h / 2 for p in y_pos], cell_block_results_df["mean_recall"],
    height=h, xerr=cell_block_results_df["std_recall"],
    color="salmon", alpha=0.8, capsize=3, label="Recall",
)
ax.set_yticks(list(y_pos))
ax.set_yticklabels(cell_block_results_df["label"])
ax.set_title("HOG Cell/Block Size Ablation — Precision & Recall — Figure 17", fontsize=13)
ax.set_xlabel("Score (5-fold CV)", fontsize=11)
ax.set_ylabel("Configuration", fontsize=11)
ax.set_xlim(0, 1)
ax.legend()
ax.grid(axis="x", linestyle="--", alpha=0.5)
plt.tight_layout()
```

### Sub-question 2 Conclusion

The ablation results above show cross-validated accuracy, AUC, F1, precision,
and recall for each cell/block size combination. The optimal configuration is
identified as the one with the highest mean F1. This finding informed the
cell/block size recommendation adopted in the thesis (Koenigstorfer, 2018).
This configuration is fixed for Sub-question 3.

## Section 3 — Sub-question 3: Orientation Bins

Tests which number of HOG orientation bins is optimal. Cell/block size is fixed
at the optimal value from Section 2, block normalisation at **L1**. The best
kernel from Section 1 (**RBF, γ = 1 / n_features**) is used.

```python
orientation_values = [6, 8, 9, 11, 13, 18]

orientation_configs = []
for ori in orientation_values:
    path = make_ablation_npz_name(FEATURES_DIR, (8, 8), (3, 3), ori)
    if not os.path.exists(path):
        print(f"WARNING: Missing ablation file, skipping: {path}")
        continue
    orientation_configs.append({
        "path":            path,
        "label":           f"ori={ori}",
        "pixels_per_cell": (8, 8),
        "cells_per_block": (3, 3),
        "orientations":    ori,
    })

print(f"Loaded {len(orientation_configs)} / {len(orientation_values)} orientation configurations.")

orientation_results_df = run_hog_ablation(
    orientation_configs, kernel="rbf", gamma=GAMMA, cv=CV_FOLDS
)
print("Orientation Ablation Results (sorted by mean F1, descending):\n")
display(orientation_results_df)

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

axes[0].barh(
    orientation_results_df["label"], orientation_results_df["mean_accuracy"],
    xerr=orientation_results_df["std_accuracy"], color="steelblue", alpha=0.8, capsize=3,
)
axes[0].set_title("Orientation Bins — Accuracy", fontsize=13)
axes[0].set_xlabel("Mean Accuracy (5-fold CV)", fontsize=11)
axes[0].set_ylabel("Orientation Bins", fontsize=11)
axes[0].set_xlim(0, 1)
axes[0].grid(axis="x", linestyle="--", alpha=0.5)

axes[1].barh(
    orientation_results_df["label"], orientation_results_df["mean_auc"],
    xerr=orientation_results_df["std_auc"], color="darkorange", alpha=0.8, capsize=3,
)
axes[1].set_title("Orientation Bins — AUC", fontsize=13)
axes[1].set_xlabel("Mean AUC (5-fold CV)", fontsize=11)
axes[1].set_ylabel("Orientation Bins", fontsize=11)
axes[1].set_xlim(0, 1)
axes[1].grid(axis="x", linestyle="--", alpha=0.5)

plt.suptitle(
    "HOG Orientation Bins Ablation — Accuracy & AUC (RBF γ=1/n_features, ppc=8, cpb=3, L1) — Figure 18",
    fontsize=13,
)
plt.tight_layout()

fig, ax = plt.subplots(figsize=(9, 6))
ax.barh(
    orientation_results_df["label"], orientation_results_df["mean_f1"],
    xerr=orientation_results_df["std_f1"], color="seagreen", alpha=0.8, capsize=3,
)
ax.set_title("HOG Orientation Bins Ablation — F1 Score — Figure 19", fontsize=13)
ax.set_xlabel("Mean F1 Score (5-fold CV)", fontsize=11)
ax.set_ylabel("Orientation Bins", fontsize=11)
ax.set_xlim(0, 1)
ax.grid(axis="x", linestyle="--", alpha=0.5)
plt.tight_layout()

fig, ax = plt.subplots(figsize=(11, 6))
y_pos = range(len(orientation_results_df))
h = 0.35

ax.barh(
    [p + h / 2 for p in y_pos], orientation_results_df["mean_precision"],
    height=h, xerr=orientation_results_df["std_precision"],
    color="cornflowerblue", alpha=0.8, capsize=3, label="Precision",
)
ax.barh(
    [p - h / 2 for p in y_pos], orientation_results_df["mean_recall"],
    height=h, xerr=orientation_results_df["std_recall"],
    color="salmon", alpha=0.8, capsize=3, label="Recall",
)
ax.set_yticks(list(y_pos))
ax.set_yticklabels(orientation_results_df["label"])
ax.set_title("HOG Orientation Bins Ablation — Precision & Recall — Figure 20", fontsize=13)
ax.set_xlabel("Score (5-fold CV)", fontsize=11)
ax.set_ylabel("Orientation Bins", fontsize=11)
ax.set_xlim(0, 1)
ax.legend()
ax.grid(axis="x", linestyle="--", alpha=0.5)
plt.tight_layout()
```

### Sub-question 3 Conclusion

The results above show cross-validated accuracy, AUC, F1, precision, and recall
for each orientation bin count. The optimal number of orientation bins is
identified as the one with the highest mean F1 and a balanced precision/recall
trade-off. Fewer bins may under-represent gradient directions, while more bins
may introduce noise without additional discriminative power.

Combined with the results from Sections 1 and 2, this completes the
identification of the optimal HOG+SVM configuration, which is used for final
model training in Section 4. These findings formed the basis for the
configuration recommended in the thesis (Koenigstorfer, 2018).

## Section 4 — Train Final Model

Using the optimal configuration identified across the three sub-questions:

| Parameter | Value |
|-----------|-------|
| Kernel | RBF |
| γ (gamma) | 1 / n_features ≈ 0.000206 |
| C | 1.0 |
| HOG block normalisation | L1 |
| Orientation bins | 11 |
| Pixels per cell | 8×8 |
| Cells per block | 3×3 |

```python
final_model = train_svm(
    X_train, y_train,
    kernel="rbf",
    C=1.0,
    gamma=GAMMA,
    probability=True,
)

train_acc = final_model.score(X_train, y_train)
print(f"Training accuracy (sanity check): {train_acc:.4f}")
```

**Note on training accuracy:** SVMs typically memorise training data when the
kernel and regularisation allow it. A near-perfect training accuracy (~1.0) is
expected and does not indicate a bug — generalisation is assessed on the holdout
test set in Section 5.

## Section 5 — Holdout Evaluation

The final model is evaluated on the held-out test set. These samples were never
seen during training or cross-validation.

```python
holdout_results = evaluate_on_holdout(final_model, X_test, y_test)

print("Holdout Test Set Results")
print("=" * 35)
for metric, label in [
    ("accuracy",  "Accuracy "),
    ("precision", "Precision"),
    ("recall",    "Recall   "),
    ("f1",        "F1 Score "),
    ("roc_auc",   "AUC      "),
]:
    print(f"  {label} : {holdout_results[metric]:.4f}")

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

plot_confusion_matrix(
    holdout_results["confusion_matrix"],
    title="Confusion Matrix",
    ax=axes[0],
)
plot_roc_curve(
    y_test, holdout_results["y_prob"],
    title="ROC Curve",
    ax=axes[1],
)
plot_precision_recall_curve(
    y_test, holdout_results["y_prob"],
    title="Precision-Recall Curve",
    ax=axes[2],
)

plt.suptitle("Final Model — Holdout Evaluation", fontsize=14)
plt.tight_layout()
```

### Comparison to Thesis Results

The thesis (Koenigstorfer, 2018, Table 18), which built on this analysis,
reports the following holdout results on the original (non-augmented) dataset:

| Metric | Thesis |
|--------|--------|
| Accuracy | ~0.70 |
| Precision | ~0.71 |
| Recall | ~0.72 |

Our results may differ due to the **4× rotational augmentation** applied during
preprocessing, which quadruples the training set size. Augmentation generally
improves generalisation, so higher holdout metrics are expected.

## Section 6 — Learning Curve

The learning curve shows how training and CV accuracy evolve as more training
data is used. This helps diagnose overfitting and assess whether additional data
would be beneficial.

```python
lc_fig = plot_learning_curve(
    X_train, y_train,
    kernel="rbf",
    gamma=GAMMA,
    cv=CV_FOLDS,
)
```

### Learning Curve Interpretation

The learning curve can be interpreted as follows:

- **Converging curves:** If training and CV scores approach each other as
  training size increases, more data would further reduce the generalisation gap.
- **High training / lower CV score:** A persistent gap indicates some degree of
  overfitting — the RBF-SVM memorises training examples. Adding data typically
  narrows this gap.
- **Plateau in CV score:** If the CV curve has plateaued, collecting additional
  data may yield diminishing returns without changes to the feature
  representation or model architecture.

Inspect the plot above to determine which regime applies to the current dataset
and whether additional data collection is likely to improve performance.

## Section 7 — Results Summary

Consolidated view of cross-validation and holdout performance for the optimal
configuration (RBF γ = 1/n_features, L1, 11 bins, 8×8 px/cell, 3×3 cells/block).

```python
# CV values for RBF γ=0.5 come from the kernel grid-search results
rbf_row = kernel_results_df[
    kernel_results_df["kernel_label"] == "RBF γ=auto"
].iloc[0]

summary_df = pd.DataFrame({
    "Metric": ["Accuracy", "AUC", "F1 Score", "Precision", "Recall"],
    "CV Mean": [
        rbf_row["mean_accuracy"], rbf_row["mean_auc"], rbf_row["mean_f1"],
        rbf_row["mean_precision"], rbf_row["mean_recall"],
    ],
    "CV Std": [
        rbf_row["std_accuracy"], rbf_row["std_auc"], rbf_row["std_f1"],
        rbf_row["std_precision"], rbf_row["std_recall"],
    ],
    "Holdout Score": [
        holdout_results["accuracy"], holdout_results["roc_auc"],
        holdout_results["f1"],       holdout_results["precision"],
        holdout_results["recall"],
    ],
})
summary_df["CV Mean"]      = summary_df["CV Mean"].map("{:.4f}".format)
summary_df["CV Std"]       = summary_df["CV Std"].map("{:.4f}".format)
summary_df["Holdout Score"] = summary_df["Holdout Score"].map("{:.4f}".format)

display(summary_df.set_index("Metric"))
```

### Discussion

**Comparison to thesis benchmark (~70% accuracy):**
The thesis (Koenigstorfer, 2018), which drew on this analysis, reports ~70%
accuracy using the same SVM configuration on the non-augmented dataset. Results
with rotational augmentation may deviate: higher accuracy indicates the
augmentation improved generalisation; lower accuracy would suggest a
distribution mismatch between augmented training images and real test images.

**Effect of rotational augmentation:**
The 4× rotational augmentation (0°, 90°, 180°, 270°) increases training
diversity without new images. For solar panels, rotational invariance is
physically justified — panels can face any cardinal direction. Augmentation
likely improves recall by ensuring the classifier sees panels at all orientations.

**Suggested next steps:**
- **More data:** The learning curve indicates additional images would further
  improve CV and holdout performance.
- **Higher resolution:** Increasing input beyond 75×75 px may capture finer
  panel boundaries.
- **CNN/ViT comparison:** A convolutional network (ResNet, EfficientNet) or
  Vision Transformer trained end-to-end would likely outperform HOG+SVM.
- **Public dataset:** The DeepSolaris/DeepGeoStat dataset
  (https://zenodo.org/records/7233404) provides a larger benchmark for direct
  comparison.
